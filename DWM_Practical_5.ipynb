```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
```

load dataset

```python
df=pd.read_csv("C:\\Users\\CG-CS-PC09\\Desktop\\DWM Practical\\wine-clustering.csv")
```

Extract Features 

```python
X=df.iloc[:,1:].values
```

Perform Agglomerative Clustering

```python
agg_clustering=AgglomerativeClustering(n_clusters=3, affinity="euclidean", linkage="ward")
```

```python
labels=agg_clustering.fit_predict(X)
```

Print Labels

```python
print("Cluster Labels:")
print(labels)
```

Create Linkage Matrix

```python
linkage_matrix=linkage(X, method="ward")
```

Plot Dendrogram

```python
plt.figure(figsize=(12,6))
dendrogram(linkage_matrix)
plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```

```python
print(X[:,0])
```

Plot Clusters in 2D

```python
plt.scatter(X[:,0], X[:,1], c=lables, s=100, cmap='rainbow')
plt.title('Agglomerative Clustering(2D)')
plt.xlabel('Alcohol')
plt.ylabel('Malik Acid')
plt.show()
```

Visualize Clusters in 3D

```python
fig=plt.figure()
ax=fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], X[:,2], c=labels)
ax.set_xlabel('Alcohol')
ax.set_ylabel('Malik Acid')
ax.set_zlabel('ash')
plt.title('Agglomerative Clustering(3D)')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(X, lables)
print(score)
```
